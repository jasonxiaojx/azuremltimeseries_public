```python
import pandas as pd
from fbprophet import Prophet
```

```python
nio_df = pd.read_csv("NIO.csv")

nio_df = nio_df.rename({"date": "ds", "close": "y"}, axis = 1)

nio_df.head()
```

```python
nio_prophet = Prophet()
nio_prophet.fit(nio_df)
```

```python
#Makes a data frame with future dates, similar to a forecasting window.
nio_future = nio_prophet.make_future_dataframe(periods = 365)
nio_future.tail()
```

```python
nio_forecast = nio_prophet.predict(nio_future)
```

```python
fig1 = nio_prophet.plot(nio_forecast)
```

```python
fig2 = nio_prophet.plot_components(nio_forecast)
```
